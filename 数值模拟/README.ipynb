# LBM_pythonic

本仓库存放数值模拟课程作业

其主要内容为基于格子玻尔兹曼法的数值模拟及其后处理

具体包括：

* 基于格子玻尔兹曼法的 **温度场** 模拟
  * 一维温度传导（Q1-3）
  * 二维温度传导（Q4-6）
    * 对于结晶潜热处理方法为 **焓方法**（Q6)
* 基于格子玻尔兹曼法的 **二维流场** 模拟（Q7-8）
  * **卡门涡街** 现象模拟（Q8）

数据后处理绘制使用到了：`matplotlib`，`plotly`，`plotly_express` 库

代码语言：`Python3`
